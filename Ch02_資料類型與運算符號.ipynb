# 資料類型與運算符號
## 資料類型
> 1. 資料類型代表記憶體的規格與大小，不可任意混合
> 2. Python的數字類型包含：整數、小數、複合數(可忽略)

```python
from IPython.display import Image
Image("./imgs/dataType.jpeg",width=500)
```

```python
#整數，小數
price = 250
print(price)
print(type(price))

price = 150.99
print(price)
print(type(price))
```

```python
#布林值類型
p1 = 5>3
p2 = 5+3 > 8
p3 = not p1 #複製p1的值，再相反
p4 = False
#計算時，True=1, False=0
p5 = True + True
#轉換時，除了0為False之外，其他皆為True
p6 = bool(0)

print(p1, p2, p3, p4, p5, p6)
```

```python
#當資料為空或0 時，判定為False
m = 0
s1 = ''
s2 = []
print(bool(m), bool(s1), bool(s2))





b1 = False
b2 = 5>3
print(m, s1, s2, b1, b2)
print(bool(m), bool(s1), bool(s2), b1, b2)
```

```python

n = -8
s11 = "Hello"
s12 = [10,20]
b2 = True
print(bool(n), bool(s11), bool(s12), b2)

print(True+False)
```

```python
a = input("請輸入任意值...")
print(bool(a))# 0 或 空-->Flase / Other-->True

y=[       ]
print(bool(y))
```

```python
# 整數+小數=?
print(price+100)

# 文字+整數=?
price = "Hello"
print(price)
print(type(price))
print(price+100)

#布林值+整數=?
price = True
print(price)
print(type(price))
print(price+100)
```

## 運算符號
> 1. 算術運算符號與賦值運算符號：得到數值
> 2. 位元運算符號不討論
> 3. 其他運算符號：得到布林值
> 4. 數值運算計算順序
>    - 小括弧優先
>    - 次方
>    - 先乘除後加減
>    - 由左至右進行


```python
from Python.display import Image
Image("",width=600)
```

```python
# 算術計算：+-*/ // ** %
a = 20
b = 3
print(a+b)
print(a-b)
print(a*b)
print(a/b)#小數除
print(a//b)#整數除
print(a%b)#求餘數
print(a**b)#求次方
```

```python
#指定運算
num=10
num+=9

print(num)
num-=8
print(num)
num*=7
print(num)
num//=6
print(num)
num%=5
print(num)
num/=4
print(num)
num**=3
print(num)
```

```python
#也適用文字
print('AB'+'CD')
s1 = 'AB'
s1+='CD'
print(s1)
print(s1*3)
```

```python
# 關係(條件)運算符號 --> 得到 True/False
a = 3
b = 5

print(a==b)
print(a!=b)
print(a>b)
print(a>=b)
print(a<b)
print(a<=b)

a1 = 'APPLE'
b1 = 'apple'
print(a1>b1)
```

```python
# 邏輯運算符號
#執行順序: not -> and -> or
#and並且:所有條件皆為True
x = 10
y = 8
print(x>y and y>=0 and x!=0)
```

```python
# or或:所有條件只要一個True
print(x>y or y==0 or x!=0)
```

```python
# not相反:把布林結果相反
print(not x>y)
print(not (x>y or y==0))
```

```python
p1=99
p2=100
print(p1>=100 and p2>=70 or p1!=p2)
print(p1>=100 and p2>=70 or p1<p2 or p1!=100)
print(p1>=100 or p2<=70 or p1<p2 and p1!=100)
print(p1>=100 or (p2<=70 or p1<p2) and p1!=100)
print(not p1>=100 or p2<=70 or p1<p2 and p1!=100)
```

```python
#in:在不在
a = 'Hello'
print('H' in a)
print('h' in a)
print('H' not in a)
print('h' not in a)
```

```python
# == 比較值，is 比較記憶體位置
a = 'Hello'
b = 'World'
print(b is a)
print(b is not a)

print(id(a))
print(id(b))

b = 'Hello'
print(b is a)
print(b is not a)
print(id(a))
print(id(b))
```
