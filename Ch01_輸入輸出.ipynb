## 輸出：print()
> 1. print()可將資料顯示在螢幕上
> 2. 可直接顯示的內容：數字、小數、變數、布林值
> 3. 需加'' 或 "" 才能顯示內容：文字
> 4. print()內的 + 符號的左右必須是相同類型的值(或變數)
> 5. print()內的，用來隔開輸出的內容，不限是否相同類型

```python
print(299)
print('a %^ 中文')
print( 150+20*3-100)
print("HE LLO~", 100, 5+3, 299.299)
print(188.25+200)
print("Hi~"+"Joe")
print(100+200+
      300)
print("0938383838")
```

```python
print(5+3)
print("ABC DEF~1#00")
print('0952380380')
print(1.41421)
print("HAHA " + "OK")
print(100, "OK", 33.88)
```

### 對輸出內容進行美化
> 1. print()預設會在資料輸出後自動換行
> 2. 若想在輸出後指定不換行，可加上：end=''
> 3. 若想在輸出中間加上換行，可加入：\n
> 4. 若想在輸出中間加上空白，可加入：\t
> 5. ，會增加一個空白，若不想有空白(或者替換別的符號)可加上：sep=""


```python
print(100, end="*******")
print(200)
print(200)
print("ABC\nEFG")
print("06\t2312312")
print("1234567812345678")
print("123\t456\t789")
print("12345\t456\t789")
print(100,"\n",200,"\n",300, sep="-->")
print("-->",100,",\n-->",200,",\n-->",300,sep="")
```

```python
print(100, "OK", 33.88, sep='')
print("OK",end='')
print(4, 5, 6, sep=' ', end='\n\n')
print("Hello\nPy\nthon")
```

```python
# 綜合練習
print("商品名稱:熱拿鐵\n單價: 85元\n數量: 3\n小計:255元")
print("-----------------------")
print("商品名稱:\tcoffee")
print("單價:\t85元")
print("數量:\t3")
print("小計:\t255元")
```

## 變數
> 1. 變數代表資料暫存的記憶體
> 2. 變數的命名方式：
>    - 第一個字：(1)大小寫英文 (2)_ (3)中文可以用，但不建議
>    - 第二字後：(1)大小寫英文 (2)_ (3)數字
>    - 不可用：空白、特殊符號、關鍵字
> 3. 變數可重覆使用，但是只會保留最後的值
> 4. 變數名稱會區分大小寫，寫錯會出錯
> 5. 變數可以單一宣告，也可以連續多個一起宣告

## 輸入
> 1. input()可接收鍵盤輸入的內容，預設以"文字"暫存
> 2. input()內可加入提示文字，提示使用者要輸入什麼
> 3. 輸入完按下 enter 可結束接收，會將資料送到記憶體，因此要指定變數來操作
> 4. 若要轉換輸入內容的類型，需要在input()外再加上轉換函數：
>    - int()：轉成整數
>    - float()：轉成小數
>    - bool()：轉成布林值

```python
x = input("請輸入任意文字...")
print(x)
```

```python
a = input("請輸入數字:")
b = input("請輸入數字:")
print(a+b)

a2 = int(a)
b2 = int(b)
print(a2+b2)

print(int(a)+int(b))
```

```python
name = input("請輸入飲料名稱:")
price = int(input("請輸入單價:"))
num = int(input("請輸入數量:"))
dis = float(input("請輸入折扣:"))
vip = bool(input("是否為會員: 1[是]  0[否]"))

print(name, price, num, dis, vip)
```

## 查詢變數的類型與記憶體位置
> 1. type(變數)：查詢變數的類型
> 2. id(變數)：查詢變數的記憶體位置

```python
a = 100
print(type(a), id(a))

a = 60
print(type(a), id(a))

a = "HAHA"
print(type(a), id(a))

w = 60
print(type(w), id(w))
```
